##Perceptron: AND gate!

```
import numpy as np
def AND(x1, x2):
    x = np.array([x1, x2]) #입력
    w = np.array([0.3, 0.1]) # 가중치
    b = -0.8 # bias(편향)
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1 
```

```
AND (0,0)
```

```
AND (1,0)
```

```
AND(0,1)
```

```
AND(1,1)
```

##Perceptron: OR gate!

```
import numpy as np
def OR(x1, x2):
    x = np.array([x1, x2]) #입력
    w = np.array([0.1, 0.5]) # 가중치
    b = -0.5 # bias(편향)
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1 
```

```
OR(0, 0)
```

```
OR(1, 0)
```

```
OR(0, 1)
```

```
OR(1, 1)
```
